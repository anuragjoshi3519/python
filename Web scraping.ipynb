Now we will be scraping images.


```python
res = requests.get("https://www.google.com/search?q=google+images&client=ubuntu&hs=Ilf&channel=fs&source=lnms&tbm=isch&sa=X&ved=0ahUKEwjKzJ2rq5zkAhXSAnIKHX88B0IQ_AUIESgB&biw=1299&bih=405")
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
soup
```

```python
soup.select('img')
```

```python
myimages = [link['src'] for link in soup.select('img')]
```

```python
myimages
```

```python
myimages = [requests.get(links).text for i,links in enumerate(myimages) if i<4]
```

```python
for i in range(len(myimages)):
    with open('image{}.jpg'.format(i),'w') as a:
        a.write(myimages[i])
```
