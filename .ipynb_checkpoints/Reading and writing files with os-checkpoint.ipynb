```python
# Now to remove a file we can use following commands
# os.unlink(path)  ->  to remove files
# os.rmdir(path)   ->  to remove empty directories
# shutil.rmtree(path)  ->  to remove all files and folders in a given location !!! most dangerous of all
```

These above commands remove the files permanently and therefore we need to be pracautious while using it.<br>
For the safe side we can move files to trash instead of permanently deleting them <br>
We need a special module for this purpose 'send2trash' module

```python
import send2trash
```

```python
send2trash.send2trash("practics.txt")
```

## Walking through a directory using os.walk()

```python
for folder, sub_folders, files in os.walk(os.getcwd()):
    print('Folder : ')
    print(folder)
    print('Sub-Folders : ')
    for sub in sub_folders:
        print(sub)
    for file in files:
        print(file)
    print('\n')
```
