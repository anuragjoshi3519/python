```python
myfile = open('message_new_pen.txt')
```

```python
myfile.seek(0)
```

```python
text = myfile.read()
```

```python
myfile.close()
```

```python
arr = text.split(', ')
```

```python
arr
```

```python
import re
```

```python
pattern = re.compile(r'[\d]+.[\d]+,-?[\d]+.[\d]+')
```

```python
match = re.findall(pattern,text)
```

```python
match
```

```python
lats = []
lons = []
```

```python
for latlon in match:
    lat,lon = latlon.split(',')
    lats.append(float(lat))
    lons.append(float(lon))
```

```python
lons
```

```python
from bokeh.io import output_file, show
```

```python
from bokeh.models import (GMapPlot,GMapOptions,ColumnDataSource, Circle, Range1d,BoxSelectTool,PanTool,WheelZoomTool)
```

```python
map_options = GMapOptions(lat=0, lng=0, zoom=3)
```

```python
plot = GMapPlot(x_range=Range1d(),y_range=Range1d(),map_options=map_options)
```

```python
plot.title.text = 'my plot'
```

```python
import getpass
```

```python
plot.api_key = getpass.getpass('API key')
```

```python
source = ColumnDataSource(data=dict(lat=lats,lon=lons))
```

```python
circle = Circle(x='lon',y='lat',size = 10,fill_color='red')
```

```python
plot.add_glyph(source,circle)
```

```python
plot.add_tools(PanTool(),WheelZoomTool(),BoxSelectTool())
```

```python
output_file('gmap_plot.html')
```

```python
show(plot)
```
